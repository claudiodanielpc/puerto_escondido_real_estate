#### RealEstate.PuertoEscondido

```python
url = "https://realestate.puerto-escondido.mx/es/propiedades/"

headers = {
    "User-Agent": "Mozilla/5.0"
}

html = requests.get(url, headers=headers).text

# =========================
# EXTRAER markers
# =========================
pattern = r'googlecode_regular_vars\.markers="(\[\[.*?\]\])"'

match = re.search(pattern, html, re.DOTALL)
markers_raw = match.group(1)

markers_raw = markers_raw.replace('\\"', '"')
markers = json.loads(markers_raw)

# =========================
# PARSEAR
# =========================
rows = []

for m in markers:
    try:
        nombre = m[0]
        lat = float(m[1])
        lon = float(m[2])

        # precio
        precio = float(m[11]) if m[11] else None

        # tipo
        tipo = m[15] if len(m) > 15 else None

        # cuartos (posición 12)
        try:
            cuartos = int(m[12])
        except:
            cuartos = None

        # baños (posición 13)
        try:
            banos = int(m[13])
        except:
            banos = None

        rows.append({
            "oferta": nombre,
            "latitud": lat,
            "longitud": lon,
            "precio": precio,
            "tipo": tipo,
            "cuartos": cuartos,
            "banos": banos
        })

    except Exception as e:
        print("error:", e)

# =========================
# DATAFRAME
# =========================
realestatepuerto = pd.DataFrame(rows)

realestatepuerto["moneda"] = "MXN"
realestatepuerto["fuente"] = "realestate_puerto_escondido"
realestatepuerto["fecha_consulta"] = datetime.now()

# limpiar nombre
realestatepuerto["oferta"] = realestatepuerto["oferta"].apply(requests.utils.unquote)

print("Total registros:", len(realestatepuerto))
realestatepuerto.head()
```

#### Century21

```python
URL = "https://century21mexico.com/busqueda/en-estado_oaxaca/en-municipio_san-pedro-mixtepec-juquila/en-colonia_san-pedro-mixtepec-juquila-puerto-escondido"

# ==============================
# DRIVER
# ==============================
options = Options()
options.add_argument("--headless=new")
options.add_argument("--window-size=1920,1080")

driver = webdriver.Chrome(options=options)
wait = WebDriverWait(driver, 20)

driver.get(URL)

# esperar a que rendericen las cards reales
wait.until(
    EC.presence_of_element_located(
        (By.CSS_SELECTOR, "div.card.mb-4.shadow")
    )
)

time.sleep(2)

# ==============================
# HTML RENDERIZADO
# ==============================
html = driver.page_source
driver.quit()

# ==============================
# PARSEAR HTML
# ==============================
soup = BeautifulSoup(html, "html.parser")

cards = soup.select("div.card.mb-4.shadow")

rows = []

for card in cards:
    # --------------------------
    # titulo
    # --------------------------
    titulo_tag = card.select_one("h5.card-title")
    titulo = titulo_tag.get_text(" ", strip=True) if titulo_tag else None

    # --------------------------
    # precio
    # --------------------------
    precio = None
    texto_card = card.get_text(" ", strip=True)

    m_precio = re.search(r"\$[\d,]+(?:\.\d+)?\s*MXN", texto_card)
    if m_precio:
        precio_txt = m_precio.group(0)
        precio = float(
            precio_txt.replace("$", "")
                      .replace(",", "")
                      .replace("MXN", "")
                      .strip()
        )

    # --------------------------
    # m² terreno y construcción
    # --------------------------
    metros_cuad_terreno = None
    metros_cuad_const = None

    m_terreno = re.search(r"([\d,]+)\s*m²T", texto_card)
    m_const   = re.search(r"([\d,]+)\s*m²C", texto_card)

    if m_terreno:
        metros_cuad_terreno = float(m_terreno.group(1).replace(",", ""))

    if m_const:
        metros_cuad_const = float(m_const.group(1).replace(",", ""))

    rows.append({
        "oferta": titulo,
        "precio": precio,
        "metros_cuad_terreno": metros_cuad_terreno,
        "metros_cuad_const": metros_cuad_const
    })

century21 = pd.DataFrame(rows)
#Añadir hora de consulta
century21["fecha_consulta"] = pd.Timestamp.now()
#Añadir fuente 
century21["fuente"] = "century21_mexico"
century21["moneda"] = "MXN"
century21.head()
```

#### Grupo Escondido PXM

```python
url = "https://www.grupoescondidopxm.mx/propiedades"

# ==============================
# DESCARGAR HTML Y LEER JSON WIX
# ==============================
html = requests.get(url).text
soup = BeautifulSoup(html, "html.parser")

script = soup.find("script", {"id": "wix-warmup-data"})
data = json.loads(script.string)

updates = data["platform"]["ssrPropsUpdates"]

# ==============================
# REPEATER UI
# ==============================
items = updates[0]["comp-maldqg5c"]["items"]

flat = {}
for d in updates[1:]:
    flat.update(d)

# ==============================
# BUSCAR DATA REAL POR item_id
# ==============================
def buscar_por_item_id(obj, item_id):
    """
    Busca recursivamente un diccionario cuyo key sea exactamente item_id.
    """
    if isinstance(obj, dict):
        for k, v in obj.items():
            if k == item_id and isinstance(v, dict):
                return v

            found = buscar_por_item_id(v, item_id)
            if found is not None:
                return found

    elif isinstance(obj, list):
        for v in obj:
            found = buscar_por_item_id(v, item_id)
            if found is not None:
                return found

    return None


# ==============================
# HELPER PARA georeferencedLocation
# ==============================
def parse_geo(geo):
    lat, lon = None, None

    if not geo:
        return lat, lon

    if isinstance(geo, str):
        try:
            lat, lon = map(float, geo.split(","))
        except:
            pass

    elif isinstance(geo, dict):
        try:
            lat = float(geo.get("latitude")) if geo.get("latitude") is not None else None
            lon = float(geo.get("longitude")) if geo.get("longitude") is not None else None
        except:
            lat, lon = None, None

    return lat, lon


# ==============================
# EXTRAER TABLA
# ==============================
rows = []

for item_id in items:

    # -------- titulo --------
    titulo_html = flat.get(f"comp-maldqg68__{item_id}", {}).get("html", "")
    titulo = BeautifulSoup(titulo_html, "html.parser").get_text(strip=True)

    # -------- precio --------
    precio_html = flat.get(f"comp-maldqg6l1__{item_id}", {}).get("html", "")
    precio_txt = BeautifulSoup(precio_html, "html.parser").get_text(strip=True)
    precio = float(re.sub(r"[^\d]", "", precio_txt)) if precio_txt else None

    # -------- data real exacta por item_id --------
    real = buscar_por_item_id(data, item_id)

    lat = lon = area = None

    if isinstance(real, dict):
        lat, lon = parse_geo(real.get("georeferencedLocation"))
        area = real.get("totalArea")

        # fallback por si viniera en location
        if (lat is None or lon is None) and isinstance(real.get("location"), dict):
            try:
                lat = float(real["location"].get("latitude")) if real["location"].get("latitude") is not None else lat
                lon = float(real["location"].get("longitude")) if real["location"].get("longitude") is not None else lon
            except:
                pass

    rows.append({
        
        "oferta": titulo,
        "precio": precio,
        "latitud": lat,
        "longitud": lon,
        "metros_cuad_terreno": area
    })

pxm = pd.DataFrame(rows)
#Añadir hora de consulta
pxm["fecha_consulta"] = pd.Timestamp.now()
#Añadir fuente 
pxm["fuente"] = "grupo escondido pxm"
pxm["moneda"] = "MXN"
```

#### Pincali

```python
def scrape_pincali():
    
    # -----------------------------
    # Consultation date
    # -----------------------------
    consultation_date = datetime.now()
    
    # -----------------------------
    # Operation mapping
    # -----------------------------
    
    all_tables = []

    # -----------------------------
    # Loop pages
    # -----------------------------
    for i in tqdm(range(1, 5), desc="Scrapeando Pincali"):
        
        # -----------------------------
        # URL
        # -----------------------------
        url = f"https://www.pincali.com/inmuebles/propiedades-en-venta?geolocation=%255B%255B14.89015257266123%252C%2520-96.7086217718429%255D%252C%2520%255B15.938592100304156%252C%2520-95.75770589646189%255D%255D&page={i}"
        
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, "html.parser")
        
        basic_data = soup.find_all("li", class_="property__component")
        
        # -----------------------------
        # Lists of data
        # -----------------------------
        name, bedrooms, bathrooms, latitude, longitude, street, region, locality, price, surface = [], [], [], [], [], [], [], [], [], []
        
        # -----------------------------
        # Loop through each property and extract data
        # -----------------------------
        for elementos in basic_data:
            
            price_tag = elementos.find("li", class_="price")
            price.append(price_tag.get_text(strip=True) if price_tag else None)
            
            name_tag = elementos.find("div", class_="title")
            name.append(name_tag.get_text(strip=True) if name_tag else None)
            
            lat = elementos.get("data-lat")
            lng = elementos.get("data-long")
            
            latitude.append(float(lat) if lat else None)
            longitude.append(float(lng) if lng else None)
            
            location_tag = elementos.find("div", class_="location")
            
            if location_tag:
                location_text = location_tag.get_text(strip=True)
                street.append(location_text)
                
                partes = [p.strip() for p in location_text.split(",")]
                locality.append(partes[-1] if len(partes) > 0 else None)
                region.append(partes[-2] if len(partes) > 1 else None)
            else:
                street.append(None)
                region.append(None)
                locality.append(None)
            
            rec = None
            bath = None
            sup = None
            
            features = elementos.find("div", class_="features")
            
            if features:
                for item in features.find_all("div"):
                    texto = item.get_text(strip=True).lower()
                    
                    numero = re.search(r"\d+", texto)
                    numero = int(numero.group()) if numero else None
                    
                    if "rec" in texto:
                        rec = numero
                    elif "bañ" in texto:
                        bath = numero
                    elif "m²" in texto or "m2" in texto:
                        sup = numero
            
            bedrooms.append(rec)
            bathrooms.append(bath)
            surface.append(sup)
            
        
        # -----------------------------
        # DataFrame for page
        # -----------------------------
        table = pd.DataFrame({
            "oferta": name,
            "precio": price,
            "metros_cuad_const": surface,
            "cuartos": bedrooms,
            "banos": bathrooms,
            "latitud": latitude,
            "longitud": longitude,
            "fecha_consulta": consultation_date
        })
        
        all_tables.append(table)

    # -----------------------------
    # Final DataFrame
    # -----------------------------
    table = pd.concat(all_tables, ignore_index=True)
    table["fuente"] = "Pincali"
    table["moneda"] = table["precio"].str.extract(r"(MXN|USD)")
    table["precio"] = (
    table["precio"]
    .str.replace(r"[^\d.,]", "", regex=True)   # quita letras y símbolos
    .str.replace(",", "")                     # quita comas de miles
    .astype(float)
)
    
    return table
```

```python
url = "https://inmobisur.mx/#desarrollos"
headers = {"User-Agent": "Mozilla/5.0"}

response = requests.get(url, headers=headers)
response.encoding = 'utf-8' 
soup = BeautifulSoup(response.text, "html.parser")

bloques = soup.find_all("div", {"data-framer-component-type": "RichTextContainer"})
data = []

for bloque in bloques:
    texto = bloque.get_text(separator="\n").strip()
    
    if "Plusvalía" in texto or "lotes" in texto:
        # Unimos todo para facilitar la búsqueda con regex
        limpio = " ".join([l.strip() for l in texto.split('\n') if l.strip()])
        
        # 1. Tiempo: Buscamos solo el número y la unidad (meses/semanas)
        # Captura: un número y la palabra que le sigue inmediatamente
        tiempo_match = re.search(r'comercialización:\s*(\d+\s*(?:meses|semanas|mes))', limpio, re.I)
        
        # 2. Metros cuadrados: Buscamos el patrón de m²
        # Algunos tienen dos valores (total | comercializado), capturamos el bloque numérico
        m2_match = re.search(r'([\d,]+\s*(?:\|\s*[\d,]+\s*)?m²)', limpio)
        
        # 3. Plusvalía, Lotes e Inversionistas (reutilizados)
        plusvalia_match = re.search(r'Plusvalía:\s*([\d\–\-%\s]+)', limpio)
        lotes_match = re.search(r'(\d+)\s*lotes', limpio)
        inv_match = re.search(r'(\d+)\s*inversionistas', limpio)

        lineas = texto.split('\n')
        registro = {
            "oferta": lineas[0].strip(),
            "ubicacion": lineas[1].strip() if len(lineas) > 1 else None,
            "metros_cuad_terreno": m2_match.group(1).replace('m²', '').strip() if m2_match else None,
            "lotes": lotes_match.group(1) if lotes_match else None,
            "inversionistas": inv_match.group(1) if inv_match else None,
            "plusvalia": plusvalia_match.group(1).strip() if plusvalia_match else None,
            "tiempo_comercializacion": tiempo_match.group(1) if tiempo_match else None
        }
        data.append(registro)

inmobisur = pd.DataFrame(data).drop_duplicates(subset=['oferta', 'ubicacion']).reset_index(drop=True)
#Añadir fuente y fecha de consulta
inmobisur["fuente"] = "Inmobisur"
inmobisur["fecha_consulta"] = pd.Timestamp.now()
inmobisur.head()
```
