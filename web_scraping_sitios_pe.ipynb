```python
#Obtener detalles adicionales de cada propiedad
def scrape_detalle(url, session):
    try:
        r = session.get(url, timeout=10)
        r.raise_for_status()
        soup = BeautifulSoup(r.text, "html.parser")

        # ── ZONA (breadcrumb) ──
        zona = None
        breadcrumbs = soup.select("ol.breadcrumb li span[itemprop='name']")
        if len(breadcrumbs) >= 2:
            zona = breadcrumbs[1].get_text(strip=True)

        # ── Bloque del agente ──
        agente_card = soup.select_one("div.card.card-agent")

        # ── ASESOR ──
        asesor = None
        if agente_card:
            asesor_tag = agente_card.select_one("h4.mt-0.mt-md-3 a")
            if asesor_tag:
                asesor = asesor_tag.get_text(strip=True)

        # ── CORREO ──
        correo = None
        if agente_card:
            for a in agente_card.find_all("a"):
                span = a.find("span", class_="icon-email")
                if span:
                    correo = a.get_text(strip=True)
                    break

        # ── TELÉFONO ──
        telefono = None
        if agente_card:
            for a in agente_card.find_all("a", href=True):
                if a["href"].startswith("tel:"):
                    telefono = a["href"].replace("tel:", "").strip()
                    break

        return zona, asesor, correo, telefono

    except Exception as e:
        print(f"  ✗ Error: {url} → {e}")
        return None, None, None, None

# ── Agregar columnas a tu tabla ya existente ──────────────────
session = requests.Session()
session.headers.update(headers)
session.cookies.update(cookies)

zonas, asesores, correos, telefonos = [], [], [], []

for i, url in enumerate(mexhome_df["link"]):
    print(f"[{i+1}/{len(mexhome_df)}] {url}")
    zona, asesor, correo, tel = scrape_detalle(url, session)
    zonas.append(zona)
    asesores.append(asesor)
    correos.append(correo)
    telefonos.append(tel)
    time.sleep(1.5)

mexhome_df["zona"]     = zonas
mexhome_df["asesor"]   = asesores
mexhome_df["correo"]   = correos
mexhome_df["contacto"] = telefonos
print("Número de columnas en mexhome_df con detalles:", mexhome_df.shape[1])

mexhome_df.head()
```

```python
url = "https://realestate.puerto-escondido.mx/es/propiedades/"

headers = {
    "User-Agent": "Mozilla/5.0"
}

html = requests.get(url, headers=headers).text

# =========================
# EXTRAER markers
# =========================
pattern = r'googlecode_regular_vars\.markers="(\[\[.*?\]\])"'

match = re.search(pattern, html, re.DOTALL)
markers_raw = match.group(1)

markers_raw = markers_raw.replace('\\"', '"')
markers = json.loads(markers_raw)

# =========================
# PARSEAR
# =========================
rows = []

for m in markers:
    try:
        nombre = m[0]
        lat = float(m[1])
        lon = float(m[2])

        # precio
        precio = float(m[11]) if m[11] else None

        # tipo
        tipo = m[15] if len(m) > 15 else None

        # cuartos (posición 12)
        try:
            cuartos = int(m[12])
        except:
            cuartos = None

        # baños (posición 13)
        try:
            banos = int(m[13])
        except:
            banos = None
        #link
        link = m[9] if m[9] else None
        #Descripción
        descripcion = m[19] if m[19] else None

        rows.append({
            "oferta": nombre,
            "latitud": lat,
            "longitud": lon,
            "precio_mxn": precio,
            "tipo": tipo,
            "cuartos": cuartos,
            "banos": banos,
            "link": link
        })

    except Exception as e:
        print("error:", e)

# =========================
# DATAFRAME
# =========================
realestatepuerto = pd.DataFrame(rows)

#realestatepuerto["moneda"] = "MXN"
realestatepuerto["fuente"] = "realestate_puerto_escondido"
realestatepuerto["fecha_consulta"] = datetime.now()

# limpiar nombre
realestatepuerto["oferta"] = realestatepuerto["oferta"].apply(requests.utils.unquote)
#Limpiar url
realestatepuerto["link"] = realestatepuerto["link"].apply(requests.utils.unquote)


print("Total registros:", len(realestatepuerto))
print("Número de columnas en realestatepuerto:", realestatepuerto.shape[1])
realestatepuerto.head()
```

```python
def scrape_realestate_puerto(url, session):
    try:
        r = session.get(url, timeout=10)
        r.raise_for_status()
        soup = BeautifulSoup(r.text, "html.parser")

        # ── ZONA ──
        zona = None
        cateqs = soup.select("div.property_categs a")
        if len(cateqs) >= 2:
            zona = cateqs[1].get_text(strip=True)

        # ── ASESOR ──
        asesor = None
        asesor_tag = soup.select_one("div.col-md-7.agent_details h3 a")
        if asesor_tag:
            asesor = asesor_tag.get_text(strip=True)

        # ── TELÉFONO ──
        telefono = None
        tel_tag = soup.select_one("div.agent_detail.agent_mobile_class a[href^='tel:']")
        if tel_tag:
            telefono = tel_tag["href"].replace("tel:", "").strip()

        # ── CORREO ──
        correo = None
        correo_tag = soup.select_one("div.agent_detail.agent_email_class a[href^='mailto:']")
        if correo_tag:
            correo = correo_tag["href"].replace("mailto:", "").strip()

        # ── METROS CUADRADOS ──
        metros = None
        for div in soup.select("div.listing_detail"):
            texto = div.get_text(separator=" ", strip=True)
            if "Lot Size" in texto:
                import re
                match = re.search(r"[\d,]+", texto)
                if match:
                    metros = match.group().replace(",", "")
                break

        return zona, asesor, telefono, correo, metros

    except Exception as e:
        print(f"  ✗ Error: {url} → {e}")
        return None, None, None, None, None


# ── Agregar columnas a tu tabla ───────────────────────────────
session = requests.Session()
session.headers.update({
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36"
})

zonas, asesores, telefonos, correos, metros_list = [], [], [], [], []

for i, url in enumerate(realestatepuerto["link"]):
    print(f"[{i+1}/{len(realestatepuerto)}] {url}")
    zona, asesor, telefono, correo, metros = scrape_realestate_puerto(url, session)
    zonas.append(zona)
    asesores.append(asesor)
    telefonos.append(telefono)
    correos.append(correo)
    metros_list.append(metros)
    time.sleep(1.5)

realestatepuerto["zona"]               = zonas
realestatepuerto["asesor"]             = asesores
realestatepuerto["contacto"]           = telefonos
realestatepuerto["correo"]             = correos
realestatepuerto["metros_cuad_terreno"] = metros_list
print("Número de columnas en realestatepuerto con detalles:", realestatepuerto.shape[1])
realestatepuerto.head()
```

#### Pincali

```python
from matplotlib.pyplot import table


def scrape_pincali():
    
    # -----------------------------
    # Consultation date
    # -----------------------------
    consultation_date = datetime.now()
    
    # -----------------------------
    # Operation mapping
    # -----------------------------
    
    all_tables = []

    # -----------------------------
    # Loop pages
    # -----------------------------
    for i in tqdm(range(1, 5), desc="Scrapeando Pincali"):
        
        # -----------------------------
        # URL
        # -----------------------------
        url = f"https://www.pincali.com/inmuebles/propiedades-en-venta?geolocation=%255B%255B14.89015257266123%252C%2520-96.7086217718429%255D%252C%2520%255B15.938592100304156%252C%2520-95.75770589646189%255D%255D&page={i}"
        
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, "html.parser")
        
        basic_data = soup.find_all("li", class_="property__component")
        
        # -----------------------------
        # Lists of data
        # -----------------------------
        name, bedrooms, bathrooms, latitude, longitude, street, region, locality, price, surface, links = [], [], [], [], [], [], [], [], [], [], []
        
        # -----------------------------
        # Loop through each property and extract data
        # -----------------------------
        for elementos in basic_data:
            
            price_tag = elementos.find("li", class_="price")
            price.append(price_tag.get_text(strip=True) if price_tag else None)
            
            link=None
            link_tag = elementos.find("a", class_="property__content property-content d-block")
            if link_tag and link_tag.has_attr("href"):
                link = "https://www.pincali.com" + link_tag["href"]
            links.append(link)

            
            name_tag = elementos.find("div", class_="title")
            name.append(name_tag.get_text(strip=True) if name_tag else None)
            
            lat = elementos.get("data-lat")
            lng = elementos.get("data-long")
            
            latitude.append(float(lat) if lat else None)
            longitude.append(float(lng) if lng else None)
            
            location_tag = elementos.find("div", class_="location")
            
            if location_tag:
                location_text = location_tag.get_text(strip=True)
                street.append(location_text)
                
                partes = [p.strip() for p in location_text.split(",")]
                locality.append(partes[-1] if len(partes) > 0 else None)
                region.append(partes[-2] if len(partes) > 1 else None)
            else:
                street.append(None)
                region.append(None)
                locality.append(None)
            
            rec = None
            bath = None
            sup = None
            
            features = elementos.find("div", class_="features")
            
            if features:
                for item in features.find_all("div"):
                    texto = item.get_text(strip=True).lower()
                    
                    numero = re.search(r"\d+", texto)
                    numero = int(numero.group()) if numero else None
                    
                    if "rec" in texto:
                        rec = numero
                    elif "bañ" in texto:
                        bath = numero
                    elif "m²" in texto or "m2" in texto:
                        sup = numero
            
            bedrooms.append(rec)
            bathrooms.append(bath)
            surface.append(sup)
            
        
        # -----------------------------
        # DataFrame for page
        # -----------------------------
        table = pd.DataFrame({
            "oferta": name,
            "precio": price,
            "metros_cuad_const": surface,
            "cuartos": bedrooms,
            "banos": bathrooms,
            "latitud": latitude,
            "longitud": longitude,
            "fecha_consulta": consultation_date,
            "link": links
        })
        
        all_tables.append(table)

    # -----------------------------
    # Final DataFrame
    # -----------------------------
    table = pd.concat(all_tables, ignore_index=True)
    table["fuente"] = "Pincali"
    table["moneda"] = table["precio"].str.extract(r"(MXN|USD)")
    table["precio"] = (
    table["precio"]
    .str.replace(r"[^\d.,]", "", regex=True)   # quita letras y símbolos
    .str.replace(",", "")                     # quita comas de miles
    .astype(float)
)
    # Después de crear la columna moneda y limpiar precio, separa en dos columnas:
    table["precio_mxn"] = table.apply(lambda r: r["precio"] if r["moneda"] == "MXN" else None, axis=1)
    table["precio_usd"] = table.apply(lambda r: r["precio"] if r["moneda"] == "USD" else None, axis=1)

    # Elimina las columnas originales
    table.drop(columns=["precio", "moneda"], inplace=True)
    
    return table
```

```python
def scrape_vivamexico(url, session):
    try:
        r = session.get(url, timeout=10)
        r.raise_for_status()
        soup = BeautifulSoup(r.text, "html.parser")

        # ── ASESOR (segundo h2.elementor-heading-title) ──
        asesor = None
        asesor_tags = soup.select("h2.elementor-heading-title")
        if len(asesor_tags) >= 2:
            asesor = asesor_tags[1].get_text(strip=True)

        # ── TELÉFONO (desde href de WhatsApp) ──
        telefono = None
        for a in soup.find_all("a", href=True):
            if "wa.me/" in a["href"]:
                match = re.search(r"wa\.me\/(\+?\d+)", a["href"])
                if match:
                    telefono = "+" + match.group(1).lstrip("+")
                break

        # ── ZONA (segunda <p> en elementor-widget-container) ──
        zona = None
        containers = soup.select("div.elementor-widget-container p")
        if len(containers) >= 2:
            texto = containers[1].get_text(strip=True)
            texto_limpio = re.sub(r'[^\w\s,]', '', texto).strip()
            zona = texto_limpio.split(",")[0].strip()

        return asesor, telefono, zona

    except Exception as e:
        print(f"  ✗ Error: {url} → {e}")
        return None, None, None


# ── Loop ─────────────────────────────────────────────────────
session = requests.Session()
session.headers.update({
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36"
})

asesores, telefonos, zonas = [], [], []

for i, url in enumerate(vivamexico["link"]):
    print(f"[{i+1}/{len(vivamexico)}] {url}")
    asesor, telefono, zona = scrape_vivamexico(url, session)
    asesores.append(asesor)
    telefonos.append(telefono)
    zonas.append(zona)
    time.sleep(1.5)

vivamexico = vivamexico.copy()
vivamexico.loc[:, "asesor"]   = asesores
vivamexico.loc[:, "contacto"] = telefonos
vivamexico.loc[:, "zona"]     = zonas
print("Número de columnas en vivamexico con detalles:", vivamexico.shape[1])
vivamexico.head()
```
